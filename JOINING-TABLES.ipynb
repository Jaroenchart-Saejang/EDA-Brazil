## customer to the geolocation

```python
customer = pd.read_csv('OLIST-DATA/olist_customers_dataset.csv')
geolocation = pd.read_csv('OLIST-DATA/olist_geolocation_dataset.csv')
```

```python
df_merged = pd.merge(customer, geolocation, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='inner')
df_merged
```

```python
geolocation_mean = geolocation.groupby('geolocation_zip_code_prefix').agg(
    geolocation_lat_mean=('geolocation_lat', 'mean'),
    geolocation_lng_mean=('geolocation_lng', 'mean'),
    geolocation_lat_std=('geolocation_lat', 'std'),
    geolocation_lng_std=('geolocation_lng', 'std')
).reset_index()

geolocation_mean
```

```python
geolocation_mean.to_csv('geolocation_mean_replace_geolocation_dataset.csv', index=False)
```
